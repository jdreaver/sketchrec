```python
def old_mhd(i_tempA, i_tempB):

    directed_distances = [0.0, 0.0]
    for i, (temp1, temp2) in enumerate([(i_tempA, i_tempB), (i_tempB, i_tempA)]):
        sum = 0.0
        for p in temp1.rasterized_points:
            sum += temp2.distance_map[p[0], p[1]]
        directed_distances[i] = sum/len(temp1.rasterized_points)
    
    return np.max(directed_distances) / (np.sqrt(2) * i_tempA.dimension)

def mhd(imageA, imageB):
    distances = [0.0, 0.0]
    directed_distances = [0.0, 0.0]
    for i, (temp1, temp2) in enumerate([(imageA, imageB), (imageB, imageA)]):
        d_sum = np.sum([temp2.distance_map[p[0], p[1]] 
                        for p in temp1.rasterized_points])
        directed_distances[i] = d_sum/len(temp1.rasterized_points)
    
    return np.max(directed_distances) / (np.sqrt(2) * imageA.dimension)

def map_mhd(imageA, imageB):
    directed_distances = [0.0, 0.0]
    for i, (temp1, temp2) in enumerate([(imageA, imageB), (imageB, imageA)]):
        d_sum = np.sum(np.array(map(lambda p: temp2.distance_map[p[0], p[1]], temp1.rasterized_points)))
        directed_distances[i] = d_sum/len(temp1.rasterized_points)
    
    return np.max(directed_distances) / (np.sqrt(2) * imageA.dimension)

def mhd_loop(imageA, imageB):
    for i in range(1000):
        distances = [0.0, 0.0]
        directed_distances = [0.0, 0.0]
        for i, (temp1, temp2) in enumerate([(imageA, imageB), (imageB, imageA)]):
            d_sum = np.sum([temp2.distance_map[p[0], p[1]] 
                            for p in temp1.rasterized_points])
            directed_distances[i] = d_sum/len(temp1.rasterized_points)
    
    return np.max(directed_distances) / (np.sqrt(2) * imageA.dimension)



def mhd_flat(imageA, imageB):
    d1 = np.sum(imageA.flat_map.take(imageB.flat_points))/imageB.num_r_points
    d2 = np.sum(imageB.flat_map.take(imageA.flat_points))/imageA.num_r_points
    return np.max([d1, d2])/(1.4142 * imageA.dimension)

def mhd_flat1(imageA, imageB):
    d1 = np.sum(imageA.flat_map.take(imageB.flat_points))/imageB.num_r_points
    d2 = np.sum(imageB.flat_map.take(imageA.flat_points))/imageA.num_r_points
    return np.max([d1, d2])/(1.4142 * 48)
    
def t(f,n):
    for i in range(n):
        f(*imgs)
```

```python
import numpy as np
from sketchrec import imageio, image_template
from sketchrec.image_template import list_classification_old

%load_ext autoreload
%autoreload 2

strokefile = 'Homework6-Problem1-text.iv'
templates = imageio.single_stroke_unlabeled_file(strokefile)
imgs = [image_template.convert_to_image(temp) for temp in templates]
temps = imgs*1
print len(temps), temps[0].dimension, temps[0].name
```

```python
def stack_1(templates):
    return np.vstack((temp.flat_map for temp in templates))
def stack_2(templates):
    return np.array([temp.flat_map for temp in templates])
def stack_3(templates):
    stack = np.zeros((len(templates), len(templates[0].flat_map)))
    for i, temp in enumerate(templates):
        stack[i] = temp.flat_map
    return stack
print np.array_equal(stack_1(temps), stack_3(temps))
%timeit stack_1(temps)
```

```python
def stack_distance_maps(templates):
    """
    Takes the flat distance maps (dim^2) of N matrices and returns and
    N by dim^2 numpy array.
    """

    return np.vstack([temp.flat_map for temp in templates])

def hauss_map_iterator(flat_map, points):
    return np.sum(flat_map.take(points))/len(points)

def vec_list_hauss(template, train):
    map_stack = stack_distance_maps(train)
    d_temp_train = np.sum(map_stack.take(template.flat_points, axis=1), 
                          axis=1)/len(template.flat_points)
    d_train_temp = np.array([hauss_map_iterator(template.flat_map, temp.flat_points) 
                            for temp in train])
    return np.maximum(d_temp_train, d_train_temp)/template.dimension
    
def list_recognition(template, train):
    hauss_distances = vec_list_hauss(template, train)
    index = np.argmin(hauss_distances)
    return (train[index].name, hauss_distances[index])

%prun a = list_recognition(temps[0], temps[1:])
print a
```
